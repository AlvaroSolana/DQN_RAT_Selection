```python
import numpy as np
import matplotlib.pyplot as plt
from channel_model import get_rate  # Asegúrate de que esto esté bien definido

def simulate_lte(N):
    distances = []
    rates_lte = []

    for _ in range(N):
        user_pos = np.random.uniform(0, 200, size=2)
        station_pos = np.random.uniform(0, 200, size=2)

        distance = np.linalg.norm(user_pos - station_pos)
        rate_lte = get_rate(user_pos, station_pos, rat_type=0)

        distances.append(distance)
        rates_lte.append(rate_lte / 1e6)  # Mbps

    # Plot
    plt.figure(figsize=(10, 6))
    plt.scatter(distances, rates_lte, alpha=0.4, s=10, label="LTE", color='blue')
    plt.xlabel("Distance (m)")
    plt.ylabel("Achievable Rate (Mbps)")
    plt.title(f"LTE: Rate vs Distance ({N} Samples)")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

simulate_lte(N=10000)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from channel_model import get_rate  # Asegúrate de que esto esté bien definido

def simulate_wifi(N):
    distances = []
    rates_wifi = []

    for _ in range(N):
        user_pos = np.random.uniform(0, 12, size=2)
        station_pos = np.random.uniform(0, 12, size=2)

        distance = np.linalg.norm(user_pos - station_pos)
        rate_wifi = get_rate(user_pos, station_pos, rat_type=1)

        distances.append(distance)
        rates_wifi.append(rate_wifi / 1e6)  # Mbps

    # Plot
    plt.figure(figsize=(10, 6))
    plt.scatter(distances, rates_wifi, alpha=0.4, s=10, label="WiFi", color='green')
    plt.xlabel("Distance (m)")
    plt.ylabel("Achievable Rate (Mbps)")
    plt.title(f"WiFi: Rate vs Distance ({N} Samples)")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Example usage
simulate_wifi(N=10000)
```
